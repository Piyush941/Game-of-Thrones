Quest Adventure Welcome to Quest Adventure! In this game, you will take on the role of an adventurer who sets out to save the kingdom. You will be presented with various choices at different stages of the quest, and your decisions will determine the outcome of the game.

Stage 1: Introduction You are greeted with a prompt asking for your name. Once you enter your name, you will be told that you are on the edge of a dark forest and asked if you would like to step into the forest.

Stage 2: Forest If you choose to step into the forest, you will be presented with the option to move in one of four directions: left, right, forward, or backward. Depending on your choice, you may encounter different obstacles or find yourself at different locations in the forest.

Stage 3: Bear If you choose to move left, you will encounter a bear. This will result in your death and the end of the game.

Stage 4: Deeper into the Forest If you choose to move right, you will continue deeper into the forest. Here, you will be presented with new obstacles and choices that will affect the outcome of your quest.

Stage 5: Final Stage You have reached the final stage of your quest. In front of you, you see a castle. To your left, there is a moat with a drawbridge. To your right, there is a steep cliff. Behind you, there is a dense forest. If you choose to go left and cross the drawbridge you will enter the castle and complete your quest, if you choose to go right, you will fall to your death, if you choose forward you will go back to the first stage and if you choose backward you will give up on your quest.

Throughout the game, you will be given feedback on your choices and the outcome of your quest. Be sure to think carefully about each decision you make and good luck on your adventure!

```python
def win():
    # win function display message when you win the game
    print('''You have successfully gathered Dragonstone and returned to 
          Nights' Watch to further protect it from Night king  
        ''')
    print("YOU HAVE WON.")

def fail(stage):
    # fail function to display message when you lose the game
    print(f"YOU HAVE FAILED AT STAGE {stage}")


def introduction():
    # introduces the story for game
    print('''
        The game is based on series Game of Thrones

        You are new at Nights' Watch, and Nights' watch
        desparately needs dragon stone hidden somewhere in forest
        to protect itself from Night King
    ''')


def stage1():
    # stage one the oath
    # returns True if passed the stage or False otherwise
    print('''
        You have sworn oath under seven gods to serve nights' watch.
        And they have chosen you to go on expedition to find dragon stone.    
    ''')
    user_answer = ""
    while user_answer.lower() not in ["yes", "no"]:
        user_answer = input(prompt="Would you like to step into the forest?\nyes/no\n prompt => ")
        if user_answer.lower() == "yes":
            print("You head into the forest. You hear crows cawwing in the distance.")
        elif user_answer.lower() == "no":
            print("You are not ready for this quest. Goodbye")
            return False # failed
        else:
            print("I didn't understand that.")

    return True


def stage2():
    # stage two the forest and bear
    # returns True if passed the stage or False otherwise
    direction = ""
    # the directions allowed
    directions = ["left", "right", "forward", "backward"]
    #using the while loop to repeat untill the required condition is met
    while direction.lower() not in directions:
        print('''
            To your left, you see a bear.
            To your right, there is more forest.
            There is a rock wall directly in front of you.
            Behind you is the forest exit. 
        ''')
        direction = input(prompt="What direction do you move?\nleft/right/forward/backward\n prompt => ")
        #conditional statements
        if direction.lower() == "left":
            print("The bear eats you. Farewell.")
            return False # Failed
        elif direction.lower() == "right":
            print("You head deeper into the forest.\n")
        elif direction.lower() == "forward":
            print("You can't scale the wall. It is too tall.\n")
            direction = ""
        elif direction.lower() == "backward":
            print("You leave the forest like coward. Goodbye.")
            return False # failed
        else:
            print("I didn't understand that.\n")

    print("You continue your journey.")
    return True

def stage3():
    # stage three the caves and dragonstone
    # returns True if passed the stage or False otherwise
    print("You see")
    print("To your left, you see a river.")
    print("To your right, there is a cave.")
    print("There is a mountain directly in front of you.")
    print("Behind you is the forest.\n")
    print("You have two chances to move")

    for chance in range(2):
        print(f"this is your chance no {chance + 1}")
        response = input(prompt="What direction do you move?\nleft/right/forward/backward\n prompt => ")
        if response.lower() == "left":
            print("You follow the river and come across a waterfall, find nothing and come back.")
        elif response.lower() == "right":
            print("You enter the cave and discover dragon stone.")
            return True # won
        elif response.lower() == "forward":
            print("You climb the mountain and reach the summit. You find nothing you come back.")
        elif response.lower() == "backward":
            print("You head back to the forest. Like coward.")
        else:
            print("I didn't understand that. You lost a chance.\n")

    return False # failed


def begin():
    # start of game

    # introduces the game story
    introduction()

    # game stages
    try:
        if stage1():
            if stage2():
                if stage3():
                    win()
                else:
                    fail(stage = 3)
            else:
                fail(stage = 2)
        else:
            fail(stage = 1)

    except Exception as e:
        print("Unexpected error happened in game. Exiting.")
        print(e)



begin()
```
